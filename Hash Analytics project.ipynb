```python
df=pd.get_dummies(df,columns=['dept'],drop_first=True)
df['salary'].replace(['low','medium','high'],[1,2,3],inplace=True)
df.drop('Emp ID',axis=1,inplace=True)
df
```

```python
pivot=pd.pivot_table(data=df,values='satisfaction_level',columns=['average_montly_hours','number_project'], index=['Left','salary'])
pivot
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(pivot).get_figure().savefig('pivottable.png',dpi=600)
```

```python
X=df.drop('Left',axis=1)
y=df['Left']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(class_weight='balanced')
dt.fit(X_train,y_train)
```

```python
predict= dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
#print(classification_report(y_test,predict))
```

```python
features= df.columns
sorted(zip(features,dt.tree_.compute_feature_importances()))
```

```python
from sklearn import tree
cn=['Existing','Left']
plt.figure(figsize=(20,12))
fig,axes=plt.subplots(nrows=1,ncols=1,dpi=400)
tree.plot_tree(dt,class_names=cn,feature_names=X.columns, filled=True);
plt.tight_layout()
```
